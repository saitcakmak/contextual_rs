Start with some violin plots to illustrate RS

```python
import numpy as np
import matplotlib.pyplot as plt
y_lims = (-3, 3)
figsize = (7, 4)

colors = ["r", "g", "b", "y"]

data1 = (
    np.random.normal(0.5, 0.8, size=10000),
    np.random.normal(-0.5, 1.2, size=10000),
    np.random.normal(0.75, 0.6, size=10000),
    np.random.normal(1.0, 1.2, size=10000),
)

fig, ax = plt.subplots(figsize=figsize)

v1 = ax.violinplot(data1, points=100, positions=np.arange(0, len(data1)),
               showmeans=True, showextrema=False, showmedians=False)
for i, b in enumerate(v1['bodies']):
    # get the center
    m = np.mean(b.get_paths()[0].vertices[:, 0])
    # modify the paths to not go further right than the center
    b.get_paths()[0].vertices[:, 0] = np.clip(b.get_paths()[0].vertices[:, 0], -np.inf, m)
    b.set_color(colors[i])

ax.set_ylim(y_lims)

fig.tight_layout()
plt.savefig("plots/rs1.pdf", dpi=300)
fig.patch.set_facecolor('xkcd:mint green')
plt.show()
```
